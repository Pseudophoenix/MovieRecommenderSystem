````python
movies.iloc[0].genres # list of dictionaries to be converted into a list of tags
````

````python
# but the given genres is a string and not a proper list object
import ast

    # list_string = "[10, 20, 'hello']"
    # actual_list = ast.literal_eval(list_string)
def convert(obj):
    obj=ast.literal_eval(obj)
    L=[]
    for i in obj:
        L.append(i['name'])
    return L
````

````python
convert(movies.iloc[0].genres)
````

````python
movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
````

````python
movies['cast']
````

````python
# for i in movies['crew']:
def director_crew(obj):
    l=[]
    i=ast.literal_eval(obj)
    for t in i:
        if t['job']=='Director':
            # print(t['name'])
            l.append(t['name'])
    return l
````

````python
movies['crew']=movies['crew'].apply(director_crew)
````

````python
movies['overview']=movies['overview'].apply(lambda x:x.split())
````

````python
movies.head()
````

We will be applying a transformation of replacing the spaces between keywords, like Johnny Depp to be converted to JohnnyDepp so that Johnny Depp and Johnny Carter don't have any similarity

````python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
````

````python
movies.head()
````

````python
movies['tags']=movies['overview']+movies['cast']+movies['crew']+movies['keywords']
````
